```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import copy
```

```python
pd.set_option('display.max_columns',63)
train = pd.read_csv('./../Data/new_data/train_.csv')
```

```python
test = pd.read_csv('./../Data/new_data/test_.csv')
```

```python
def striper(sentence):
    return sentence.strip()

province = ['Eastern Cape', 'Free State', 'Gauteng', 'KwaZulu-Natal', 'Limpopo', 'Mpumalanga', 'Northern Cape', 'North West','Western Cape']

def add_variables(sentence):
    temp = sentence.split(',')
    temp = map(striper,temp)
    for p in temp:
        if p in province:
            adress = p
            return adress
```

```python
train['province'] = np.vectorize(add_variables)(train['address'])
train.shape
```

```python
test['province'] = np.vectorize(add_variables)(test['address'])
test.shape
```

```python
data = train.append(test)
data['province'].value_counts()
```

```python
## features transform
def transform(data):
    data['pw1'] = 1 - (data['pw_00'] + data['pw_06'])    
    data['rich'] = (data['car_00']+data['lgt_00']+data['pw_00'])/3
    data['psa_car'] = data['psa_00']/(data['car_00'] + data['car_01'])
    data['latlon'] = abs(data['lat']) + abs(data['lon'])
    #"""data['stv_00']+data['psa_00']+"""
    #data['poor'] = 1 - data['rich']
    #data['nl_individuals'] = data['NL'] / data['NL'].max() * data['total_individuals']
    #data['dw_00_'] = data['dw_00'] + data['dw_01']+ data['dw_02']+ data['dw_03']+ data['dw_04']+ data['dw_05']+ data['dw_06']+ data['dw_07']+ data['dw_08']+ data['dw_09']+ data['dw_11']
    #data['pg_01_'] = data['pg_01'] + data['pg_02'] + data['pg_03'] + data['pg_04']
```

```python
## features removing

def remove(data):
    data.drop(['total_individuals'],axis=1,inplace=True)
    data.drop(['total_households'],axis=1,inplace=True)
    #data.drop(['nl_individuals'],axis=1,inplace=True)
    #data.drop(['nl_households'],axis=1,inplace=True)
    #data.drop(['NL'],axis=1,inplace=True)
    data.drop(['address'],axis=1,inplace=True)
    #data.drop(['province'],axis=1,inplace=True)
    #data.drop(['lat'],axis=1,inplace=True)
    #data.drop(['lon'],axis=1,inplace=True)
    data.drop(['lgt_00'],axis=1,inplace=True)
    #data.drop(['dw_00','dw_01','dw_02','dw_03','dw_04','dw_05','dw_06','dw_07','dw_08','dw_09','dw_11'],axis=1,inplace=True)
    #data.drop(['stv_00','car_00','psa_00','stv_01','car_01','psa_01'],axis=1,inplace=True)
    data.drop(['psa_02','psa_03'],axis=1,inplace=True)
    data.drop(['lan_02','lan_03','lan_04','lan_05','lan_08','lan_07'],axis=1,inplace=True)
    data.drop(['dw_00','dw_02', 'dw_06'],axis=1,inplace=True)
    data.drop(['pw_01','pw_02','pw_03','pw_04','pw_05','pw_07','pw_08'],axis=1,inplace=True)
```

```python
## remove data without importance
def remove_(data):
    for col in data.select_dtypes(int):
        if col != 'province':
            data.drop([col],axis=1,inplace=True)
```

```python
def encoder(data):
    code = {
        'KwaZulu-Natal':8,
        'Eastern Cape':7,
        'Gauteng':6,
        'Limpopo':5,
        'Western Cape':4,
        'Mpumalanga':3,
        'North West':2,
        'Free State':1,
        'Northern Cape':0
    }
    data.loc[:,'province'] = data['province'].map(code)
    
```

```python
def preprocessing(data):
    encoder(data)
    transform(data)
    remove(data)
    remove_(data)  
```

```python
preprocessing(train)
preprocessing(test)
```
